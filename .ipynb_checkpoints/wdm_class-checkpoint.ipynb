```python
class HDF5KeyImageDataset(Dataset):
    def __init__(self, cdm_path, wdm_path, transform=None):
        self.transform = transform

        self.data_sources = []  # List of (file_type, key_name, sub_index_if_needed)
        
        # Load CDM keys
        self.cdm_file = h5py.File(cdm_path, 'r')
        for key in self.cdm_file.keys():
            self.data_sources.append(('cdm', key, None))

        # Load WDM keys
        self.wdm_file = h5py.File(wdm_path, 'r')
        for key in self.wdm_file.keys():
            self.data_sources.append(('wdm', key, None))


    def __len__(self):
        return len(self.data_sources)

    def __getitem__(self, idx):
        source, key, sub_idx = self.data_sources[idx]
        h5_file = self.cdm_file if source == 'cdm' else self.wdm_file
        label = 0 if source == 'cdm' else 1
    
        arr = h5_file[key]
        img = arr[:] if sub_idx is None else arr[sub_idx]
    
        # Clean + normalize
        img = np.nan_to_num(img, nan=0.0, posinf=1.0, neginf=0.0)
        img = np.clip(img, 0, None)  # Ensure non-negative
        img = img.astype(np.float32)
    
        if img.max() > 0:
            img = img / img.max()  # Normalize to [0, 1]
    
        # Convert to torch tensor and add channel dim
        img = torch.from_numpy(img).unsqueeze(0)  # shape: [1, H, W]
    
        if self.transform:
            img = self.transform(img)
    
        return img, torch.tensor(label, dtype=torch.float32)


    def __del__(self):
        self.cdm_file.close()
        self.wdm_file.close()
```

```python
from torch.utils.data import ConcatDataset

def load_multiple_hdf5_datasets(indices, transform=None):
    datasets = []
    for idx in indices:
        cdm_path = f'/n/netscratch/iaifi_lab/Lab/ccuestalazaro/DREAMS/Images/CDM/MW_zooms/box_{idx}/CDM/Galaxy_{idx}.hdf5'
        wdm_path = f'/n/netscratch/iaifi_lab/Lab/ccuestalazaro/DREAMS/Images/WDM/MW_zooms/box_{idx}/CDM/Galaxy_{idx}.hdf5'
        if os.path.exists(cdm_path) and os.path.exists(wdm_path):
            ds = HDF5KeyImageDataset(cdm_path, wdm_path, transform=transform)
            datasets.append(ds)
        else:
            print(f"Skipping missing: {cdm_path} or {wdm_path}")
    return ConcatDataset(datasets)
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import random

class TensorAugment:
    def __init__(self, size=(64, 64), mean=0.5, std=0.5, p_flip=0.5, p_rot=0.5, noise_std=0.01):
        self.size = size
        self.mean = mean
        self.std = std
        self.p_flip = p_flip
        self.p_rot = p_rot
        self.noise_std = noise_std
    
    def __call__(self, img):  # img: torch.Tensor of shape [1, H, W] or [H, W]
        # Handle different input shapes
        if img.dim() == 2:  # [H, W]
            img = img.unsqueeze(0)  # Add channel dimension -> [1, H, W]
        elif img.dim() == 3 and img.shape[0] != 1:  # [C, H, W] where C != 1
            img = img[0:1]  # Take first channel -> [1, H, W]
        
        # Resize - add batch dimension for interpolate
        img = F.interpolate(img.unsqueeze(0), size=self.size, mode='bilinear', align_corners=False)
        img = img.squeeze(0)  # Remove batch dimension -> [1, 64, 64]
        
        # Horizontal/vertical flips
        if random.random() < self.p_flip:
            img = torch.flip(img, dims=[2])  # horizontal flip
        if random.random() < self.p_flip:
            img = torch.flip(img, dims=[1])  # vertical flip
        
        # Rotation (by multiples of 90째)
        if random.random() < self.p_rot:
            k = random.choice([1, 2, 3])  # 90째, 180째, 270째
            img = torch.rot90(img, k=k, dims=[1, 2])
        
        # Add Gaussian noise
        noise = torch.randn_like(img) * self.noise_std
        img = img + noise
        
        # Normalize
        img = (img - self.mean) / self.std
        
        return img

# Create transform
transform = TensorAugment(
    size=(64, 64),
    mean=0.5,
    std=0.5,
    p_flip=0.5,
    p_rot=0.5,
    noise_std=0.01
)
```

```python
WDM_mass=[]
with open('WDM_TNG_MW_SB4.txt','r') as f:
    for i, line in enumerate(f.readlines()[1:]):
        WDM_mass.append([i,float(line.strip().split(' ')[0])])

WDM_mass = np.array(WDM_mass)

k = 10
# sorted_mass = WDM_mass[WDM_mass[:,1].argsort()]
# all_indices = sorted_mass[:k,0].astype(int).tolist()

all_indices = random.sample(range(1024),k)
```

```python
# Set random seed for reproducibility
import random
import torch
import numpy as np

def set_seed(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed(seed)
        torch.cuda.manual_seed_all(seed)

set_seed(42)  # Set seed for reproducible results

# Shuffle indices to randomize before splitting
random.shuffle(all_indices)

# Define split ratios (60% train, 20% val, 20% test)
train_ratio, val_ratio, test_ratio = 0.6, 0.2, 0.2
total_samples = len(all_indices)

# Calculate split indices
train_end = int(train_ratio * total_samples)
val_end = int((train_ratio + val_ratio) * total_samples)

# Create index splits
train_indices = all_indices[:train_end]
val_indices = all_indices[train_end:val_end]
test_indices = all_indices[val_end:]

# Print split information
print(f"Dataset split:")
print(f"Total samples: {total_samples} sets")
print(f"Train: {len(train_indices)} sets ({len(train_indices)/total_samples*100:.1f}%)")
print(f"Validation: {len(val_indices)} sets ({len(val_indices)/total_samples*100:.1f}%)")
print(f"Test: {len(test_indices)} sets ({len(test_indices)/total_samples*100:.1f}%)")

# Create datasets
# Training set: WITH augmentation
train_dataset = load_multiple_hdf5_datasets(train_indices, transform=transform)

# Validation and test sets: WITHOUT augmentation
val_dataset = load_multiple_hdf5_datasets(val_indices, transform=None)
test_dataset = load_multiple_hdf5_datasets(test_indices, transform=None)

# Create data loaders
# Use num_workers=1 or 0 based on system capabilities
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True, num_workers=1, pin_memory=True)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False, num_workers=1, pin_memory=True)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False, num_workers=1, pin_memory=True)

print(f"\nData loaders created:")
print(f"Train loader: {len(train_loader)} batches")
print(f"Val loader: {len(val_loader)} batches") 
print(f"Test loader: {len(test_loader)} batches")
```

```python
# Robust CNN that adapts to any input size
class AdaptiveCNN(nn.Module):
    def __init__(self, dropout_conv=0.1, dropout_fc=0.3):
        super(AdaptiveCNN, self).__init__()
        
        self.conv_layers = nn.Sequential(
            # First conv block
            nn.Conv2d(1, 16, 3, padding=1),
            nn.BatchNorm2d(16),
            nn.ReLU(),
            nn.MaxPool2d(2),
            nn.Dropout2d(dropout_conv),
            
            # Second conv block
            nn.Conv2d(16, 32, 3, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.MaxPool2d(2),
            nn.Dropout2d(dropout_conv),
            
            # Third conv block
            nn.Conv2d(32, 64, 3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.MaxPool2d(2),
            nn.Dropout2d(dropout_conv),
        )
        
        # Use adaptive pooling to ensure consistent output size
        self.adaptive_pool = nn.AdaptiveAvgPool2d((4, 4))  # Always outputs 4x4
        
        self.fc = nn.Sequential(
            nn.Flatten(),
            nn.Linear(64 * 4 * 4, 128),  # 64 channels * 4 * 4 = 1024
            nn.BatchNorm1d(128),
            nn.ReLU(),
            nn.Dropout(dropout_fc),
            nn.Linear(128, 64),
            nn.ReLU(),
            nn.Dropout(dropout_fc),
            nn.Linear(64, 1)
        )
    
    def forward(self, x):
        x = self.conv_layers(x)
        x = self.adaptive_pool(x)  # Ensure consistent size
        return self.fc(x)

# Alternative: Fixed size CNN (if you're sure about 64x64 input)
# class SimpleCNN(nn.Module):
#     def __init__(self, dropout_conv=0.1, dropout_fc=0.3):
#         super(SimpleCNN, self).__init__()
        
#         self.conv_layers = nn.Sequential(
#             nn.Conv2d(1, 16, 3, padding=1),
#             nn.BatchNorm2d(16),
#             nn.ReLU(),
#             nn.MaxPool2d(2),  # 64->32
#             nn.Dropout2d(dropout_conv),
            
#             nn.Conv2d(16, 32, 3, padding=1),
#             nn.BatchNorm2d(32),
#             nn.ReLU(),
#             nn.MaxPool2d(2),  # 32->16
#             nn.Dropout2d(dropout_conv),
            
#             nn.Conv2d(32, 64, 3, padding=1),
#             nn.BatchNorm2d(64),
#             nn.ReLU(),
#             nn.MaxPool2d(2),  # 16->8
#             nn.Dropout2d(dropout_conv),
#         )
        
#         # For 64x64 input: 64->32->16->8, so 64*8*8=4096
#         self.fc = nn.Sequential(
#             nn.Flatten(),
#             nn.Linear(64 * 8 * 8, 128),  # 4096 -> 128
#             nn.BatchNorm1d(128),
#             nn.ReLU(),
#             nn.Dropout(dropout_fc),
#             nn.Linear(128, 64),
#             nn.ReLU(),
#             nn.Dropout(dropout_fc),
#             nn.Linear(64, 1)
#         )
    
#     def forward(self, x):
#         x = self.conv_layers(x)
#         return self.fc(x)

# Debug function to check dimensions
def debug_model_dimensions(model, input_size=(1, 64, 64)):
    model.eval()
    with torch.no_grad():
        x = torch.randn(1, *input_size)  # Batch size 1
        print(f"Input shape: {x.shape}")
        
        if hasattr(model, 'conv_layers'):
            x_conv = model.conv_layers(x)
            print(f"After conv layers: {x_conv.shape}")
            
            if hasattr(model, 'adaptive_pool'):
                x_pool = model.adaptive_pool(x_conv)
                print(f"After adaptive pool: {x_pool.shape}")
                x_flat = x_pool.flatten(1)
            else:
                x_flat = x_conv.flatten(1)
            
            print(f"After flatten: {x_flat.shape}")
            print(f"Expected linear input size: {x_flat.shape[1]}")


# Usage examples:
# For normal training:
# model = SimpleCNN(dropout_conv=0.1, dropout_fc=0.3)

# If still overfitting, try lower dropout:
# model = SimpleCNN(dropout_conv=0.05, dropout_fc=0.2)

# If severe overfitting, use the simpler model:
# model = SimplerCNN()
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = AdaptiveCNN(dropout_conv=0, dropout_fc=0.2).to(device)
criterion = nn.BCEWithLogitsLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=5e-4, weight_decay=1e-3)

# Initialize scheduler ONCE before training loop
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, patience=5, factor=0.5)

EPOCHS = 60
loss_history = []
accuracy_history = []
val_loss_history = []
val_accuracy_history = []

# Early stopping variables
best_val_loss = float('inf')
patience = 10
patience_counter = 0

print("Starting training...")

for epoch in range(EPOCHS):
    # TRAINING PHASE
    model.train()
    total_loss, correct = 0.0, 0
    
    for images, labels in train_loader:
        images = images.to(device)
        labels = labels.to(device).unsqueeze(1)
        
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        
        total_loss += loss.item() * images.size(0)
        preds = (torch.sigmoid(outputs) > 0.5).float()  # Apply sigmoid for prediction
        correct += (preds == labels).sum().item()
    
    # Calculate training metrics
    epoch_loss = total_loss / len(train_dataset)
    epoch_acc = correct / len(train_dataset)
    loss_history.append(epoch_loss)
    accuracy_history.append(epoch_acc)
    
    # VALIDATION PHASE
    model.eval()
    val_loss, val_correct = 0.0, 0
    
    with torch.no_grad():
        for images, labels in val_loader:
            images = images.to(device)
            labels = labels.to(device).unsqueeze(1)
            
            outputs = model(images)
            loss = criterion(outputs, labels)
            
            val_loss += loss.item() * images.size(0)
            preds = (torch.sigmoid(outputs) > 0.5).float()  # Apply sigmoid for prediction
            val_correct += (preds == labels).sum().item()
    
    # Calculate validation metrics
    val_epoch_loss = val_loss / len(val_dataset)
    val_epoch_acc = val_correct / len(val_dataset)
    val_loss_history.append(val_epoch_loss)
    val_accuracy_history.append(val_epoch_acc)
    
    # Print metrics
    print(f"Epoch {epoch+1}/{EPOCHS} | Train Loss: {epoch_loss:.4f} | Train Acc: {epoch_acc:.4f} | Val Loss: {val_epoch_loss:.4f} | Val Acc: {val_epoch_acc:.4f}")
    
    # Learning rate scheduling
    scheduler.step(val_epoch_loss)
    
    # Early stopping logic
    if val_epoch_loss < best_val_loss:
        best_val_loss = val_epoch_loss
        patience_counter = 0
        torch.save(model.state_dict(), 'best_model.pth')
        print(f"New best model saved with validation loss: {val_epoch_loss:.4f}")
    else:
        patience_counter += 1
        if patience_counter >= patience:
            print(f"Early stopping at epoch {epoch+1} - No improvement for {patience} epochs")
            break

# Load the best model for final save
model.load_state_dict(torch.load('best_model.pth',weights_only=True))
torch.save(model.state_dict(), 'cdm_wdm_classifier_hdf5.pth')
print("Training completed. Best model saved as 'cdm_wdm_classifier_hdf5.pth'")
```

```python
def compute_saliency_map(model, image, label):
    model.eval()

    # Ensure image shape: [1, 1, H, W] and type: float32
    if image.dim() == 2:  # shape [H, W]
        image = image.unsqueeze(0).unsqueeze(0)  # -> [1, 1, H, W]
    elif image.dim() == 3:  # shape [1, H, W]
        image = image.unsqueeze(0)  # -> [1, 1, H, W]

    image = image.to(device).float()
    image.requires_grad_()

    # Label: ensure it's a float32 tensor with shape [1, 1]
    label_tensor = torch.tensor([[float(label)]], dtype=torch.float32, device=device)

    # Forward pass
    output = model(image)  # shape: [1, 1]
    loss = nn.BCEWithLogitsLoss()(output, label_tensor)
    loss.backward()

    # Saliency is the absolute value of the input gradient
    saliency = image.grad.abs().squeeze().detach().cpu()  # shape: [H, W]
    return saliency


    
import matplotlib.pyplot as plt

def show_saliency(image, saliency, title='Saliency Map', label=0):
    """
    Show input image and its corresponding saliency map side by side.
    """
    fig, ax = plt.subplots(1, 2, figsize=(10, 4))

    # Input image
    ax[0].imshow(image.squeeze().cpu(), cmap='gray')
    ax[0].set_title('WDM Input Image' if label else 'CDM Input Image')
    ax[0].axis('off')

    # Saliency map
    ax[1].imshow(saliency, cmap='inferno')
    ax[1].set_title(title)
    ax[1].axis('off')

    plt.tight_layout()
    plt.savefig(f'Saliency Map {int(label)}.png', bbox_inches='tight')
    plt.show()

```

```python
def extract_flattened(model, image):
    image = image.unsqueeze(0).to(device)
    with torch.no_grad():
        features = model(image)  # <- fix here
        flat = features.view(1, -1).cpu().numpy()
    return flat
```

```python
import h5py
import numpy as np
import matplotlib.pyplot as plt

# Load all images from all keys
all_images = []

def demo(dmtype='CDM'):
    with h5py.File('/n/netscratch/iaifi_lab/Lab/ccuestalazaro/DREAMS/Images/WDM/MW_zooms/box_37/CDM/Galaxy_37.hdf5', 'r') as f:
        for key in f.keys():
            all_images.append(f[key][:])
    
    # Choose how many to display
    num_images = min(64, len(all_images))
    grid_size = int(np.ceil(np.sqrt(num_images)))
    
    # Plot
    fig, axes = plt.subplots(grid_size, grid_size, figsize=(12, 12))
    axes = axes.flatten()
    
    for i in range(grid_size * grid_size):
        ax = axes[i]
        if i < num_images:
            img = np.log1p(all_images[i])  # log scale for visibility
            ax.imshow(img, cmap='magma')
        ax.axis('off')
    
    plt.tight_layout()
    plt.show()

demo(dmtype='CDM')
```
